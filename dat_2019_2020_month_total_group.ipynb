```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import calendar
```

```python
dat=pd.read_csv("D:/Second_paper_data/COVID-EMDA-master/data_release/caiso/dat_2019_2020_month_total_group.csv")
```

```python
dat.head()
```

```python
dat1=dat[dat['year']==2019]
dat2=dat[dat['year']==2020]

dat1.sort_values(by=['month'], inplace=True)
dat2.sort_values(by=['month'], inplace=True)
ax = plt.figure(figsize=(6, 4), dpi=80).add_subplot(111)

ax.plot(dat1['month'], dat1['Month_total'], linewidth = 1,marker='o',markersize=4,label="2019")
ax.plot(dat2['month'], dat2['Month_total'], linewidth = 1,marker='o',markersize=4,label="2020")

ticklabels = ['January','February','March','April','May','June','July','August','September','October','November','December']
ax.set_xticks(np.arange(1,13))
ax.set_xticklabels(ticklabels) #add monthlabels to the xaxis
ax.tick_params(labelrotation=40)
ax.legend( loc = 'best',prop={ 'size':10})
#ax.legend(pt.columns.tolist(), loc='center left', bbox_to_anchor=(1, .5)) #add the column names as legend.
#plt.tight_layout(rect=[0, 0, 0.85, 1])
plt.gcf().axes[0].yaxis.get_major_formatter().set_scientific(False)
plt.xlabel('Month',fontdict={'size':12})
plt.ylabel('Electricity consumption (kWh)',fontsize=12)
plt.show()
```
