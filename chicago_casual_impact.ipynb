```python
import pandas as pd
import numpy as np
import matplotlib.dates as mdates
import matplotlib.pyplot as plt
import seaborn as sns
import collections
from datetime import datetime
%matplotlib inline
```

```python
dateparse = lambda x: datetime.strptime(x, '%d/%m/%Y')
df=pd.read_csv('D:/Second_paper_data/COVID-EMDA-master/data_release/pjm/chicago_casual_result.csv', parse_dates=['date'],date_parser=dateparse)
```

```python
df.head()
```

```python
df1=df[(df['date'] >= '2019-01-01') & (df['date'] <= '2020-04-24')]
```

```python
colors = sns.color_palette()
c1, c2, c3, c4, c5, c6,c7= colors[0], colors[1], colors[2], colors[3], colors[4], colors[5],colors[6]
fig, ax = plt.subplots(figsize=(10,5))
ax.plot(df1['date'], df1['response'],label='Actual',c=c2,lw=2,alpha=1,linestyle='dashed',marker='o',markersize=2)
ax.plot(df1['date'], df1['point.pred'],label='Prediction',c=c5,lw=2,alpha=1,linestyle='dotted',marker='o',markersize=2)
ax.xaxis.set_major_locator(mdates.DayLocator(interval=30))
ax.set_ylabel("Daily Load(kWh)")
ax.set_xlabel("Date")
ax.axvline('2020-03-14', color='red', linestyle='--', label='Begin of Lockdown')
plt.setp(ax.xaxis.get_majorticklabels(), rotation=45)
#ax.set_ylim(40000,90000)
ax.legend(loc='upper left')
#ax.set(title='New York Daily Load')
```
