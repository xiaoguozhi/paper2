```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import calendar
```

```python
dat=pd.read_csv("D:/Second_paper_data/COVID-EMDA-master/data_release/caiso/dat_2019_2020_week_year_total.csv")
```

```python
dat.head()
```

```python
d = dict(zip(list(calendar.day_name),range(7)))
```

```python
dat['DayOfTheWeek'] = dat['week'].map(d)
```

```python
dat
```

```python
dat1=dat[dat['year']==2019]
dat2=dat[dat['year']==2020]

dat1.sort_values(by=['DayOfTheWeek'], inplace=True)
dat2.sort_values(by=['DayOfTheWeek'], inplace=True)
ax = plt.figure(figsize=(6, 4), dpi=80).add_subplot(111)

ax.plot(dat1['DayOfTheWeek'], dat1['Week_total'], linewidth = 1,marker='o',markersize=4,label="2019")
ax.plot(dat2['DayOfTheWeek'], dat2['Week_total'], linewidth = 1,marker='o',markersize=4,label="2020")

ticklabels = ['Monday','Tuesday','Thursday','Wednesday','Friday','Saturday','Sunday']
ax.set_xticks(np.arange(0,7))
ax.set_xticklabels(ticklabels) #add monthlabels to the xaxis
ax.tick_params(labelrotation=40)
ax.legend( loc = 'best',prop={ 'size':10})
#ax.legend(pt.columns.tolist(), loc='center left', bbox_to_anchor=(1, .5)) #add the column names as legend.
#plt.tight_layout(rect=[0, 0, 0.85, 1])
plt.gcf().axes[0].yaxis.get_major_formatter().set_scientific(False)
plt.xlabel('Day of week',fontdict={'size':12})
plt.ylabel('Electricity consumption (kWh)',fontsize=12)
plt.show()
```
